```python
import pandas as pd
data=pd.read_csv('../input/google-play-storecsv/googleplaystore.csv')
```

```python
#清理10472後,重新計算
condition=data['Rating']<=5
data=data[condition]
print(data['Rating']) #ex對於rating有興趣
print("平均數:",data['Rating'].mean()) 
print("中位數",data['Rating'].median())
print("前1000個app的平均",data['Rating'].nlargest(1000).mean())
```

```python
#分析資料,安裝數量
print("平均數",data['Installs'].mean())
```

## 發現安裝數量是字串呈現,因此對他做平均就出事拉

### 使用pandas 中的to_numeric

```python
data['Installs']=pd.to_numeric(data['Installs'])
```

### 發現無法處理,因為是10,000+ 裡面有, +不是數字的東西要處理掉

```python
data['Installs']=pd.to_numeric(data['Installs'].str.replace("[,+]","").replace("Free","")) #wow 好讚
```

```python
print("平均數",data['Installs'].mean()) #經過處理後就可以進行計算下載量的average
```

```python
#安裝數量>10w
condition=data['Installs']>100000
data=data[condition]
print("安裝數量>10w",data[condition].shape[0])
```

```python
#基於資料應用:關鍵字搜尋應用程式名稱
keyword=input("請輸入關鍵字")
condition=data['App'].str.contains(keyword,case=False)
print('包含關鍵字的應用程式數量',data[condition].shape[0])
```
